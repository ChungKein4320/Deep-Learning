## **ACTIVATION**

### **Sigmoid**

```python
# nh ngha hm sigmoid 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# o hm sigmoid
def sigmoid_deriv(x):
    return  sigmoid(x) * (1 - sigmoid(x))
```

### **Tanh**

![Code_Generated_Image (1).png](<attachment:Code_Generated_Image (1).png>)

```python
# nh ngha hm tanh 
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) # or np.tanh(x)
```

### **ReLU**

![Code_Generated_Image (2).png](<attachment:Code_Generated_Image (2).png>)

```python
# nh ngha hm ReLu
def relu(x):
    return np.maximum(0, x)
```
